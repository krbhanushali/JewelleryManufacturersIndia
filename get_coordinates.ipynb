```
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt

from geopy.geocoders import Nominatim

data = pd.read_csv('data.csv', header=0)

all_data = pd.DataFrame(columns=['name','address','products','precious_metal','gemstone', 'artificial', 'misc','loc_data' ,'co-ordinates'])
all_data['name'] = data['name']
all_data['address'] = data['address']
all_data['products'] = data['product']

desiredwords = ['Jewellery']
precious_stones = ['Ruby', 'Solitaire', 'Tanzanite', 'Diamond','Semi-Precious',  'Sapphire', 'Gemstone', 'Emerald' ,'Pearl', 'Gem']
precious_metal = ['Gold', 'Silver', 'Platinum', 'Sterling']
artificial = ['Sozo', 'CZ', 'Costume' ,'Imitation', 'Synthetic', 'Artificial']

metal = []
gem = []
arti = []
misc = []
for i in range(len(all_data['products'])):
  all_data['products'][i] = all_data['products'][i].strip('][').split(', ')
  all_data['products'][i] = [x for x in all_data['products'][i] if any(word in x for word in desiredwords)]
  metal.append(len([x for x in all_data['products'][i] if any(word in x for word in precious_metal)]))
  gem.append(len([x for x in all_data['products'][i] if any(word in x for word in precious_stones)]))
  arti.append(len([x for x in all_data['products'][i] if any(word in x for word in artificial)]))
  misc.append(len(all_data['products'][i])- metal[i]- gem[i]- arti[i])

all_data['precious_metal'] = metal
all_data['gemstone'] = gem
all_data['artificial'] = arti
all_data['misc'] = misc
```

```
addresses = [] 
for i in range(len(all_data['address'])):
   curr = all_data['address'][i]
   
   curr = curr.split(',')
   if(len(curr)>=3): 
     curr = curr[-3: ]
   
   curr = ','.join(curr)
   curr = curr.replace('-', " , ")
   curr = curr.strip('-|/!@#$%^&*"').split(',')
   curr = [c for c in curr if (c!= "" and c!= " ")]

   addresses.append(','.join(curr))

all_data['address'] = addresses
```

```
from geopy.extra.rate_limiter import RateLimiter
locator = Nominatim(user_agent= "krishnabhanushali123@gmail.com")
geocode = RateLimiter(locator.geocode, min_delay_seconds=1)
all_data['loc_data'] = all_data['address'].apply(geocode)

all_data.to_csv('data_with_loc.csv')
```

```
all_data['co-ordinates'] = all_data['loc_data'].apply(lambda loc: tuple(loc.point) if loc else (None,None, None))
all_data[['X','Y', 'Alt']] =  pd.DataFrame(all_data['co-ordinates'].tolist(), index=all_data.index)
all_data.to_csv('data_with_loc.csv')
```
