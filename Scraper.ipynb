```
import sys
sys.path.insert(0,'/usr/lib/chromium-browser/chromedriver')
from selenium import webdriver
chrome_options = webdriver.ChromeOptions()
chrome_options.add_argument('--headless')
chrome_options.add_argument('--no-sandbox')
chrome_options.add_argument('--disable-dev-shm-usage')
wd = webdriver.Chrome('chromedriver',chrome_options=chrome_options)

from bs4 import BeautifulSoup
import time
wd.get("http://directory.indianjeweller.in/Search/Jewellery")
wd.maximize_window()
wd.refresh()

explicit_wait_timeout_secs = 10

def wait_for_element_presence(driver, find_type, web_element):
    return selenium.webdriver.support.ui.WebDriverWait(driver, explicit_wait_timeout_secs).until(EC.presence_of_element_located((find_type, web_element)))


def wait_for_element_clickable(driver, find_type, web_element):
    return selenium.webdriver.support.ui.WebDriverWait(driver, explicit_wait_timeout_secs).until(EC.element_to_be_clickable((find_type, web_element)))


def scroll_element_into_view(driver, element):
    """Scroll element into view"""
    y = element.location['y']
    driver.execute_script('window.scrollTo(0, {0})'.format(y))
    
search_button = wd.find_element_by_id('btnSearch') #search button
wd.execute_script("arguments[0].click();", search_button)
time.sleep(10)

AtoZorder = wd.find_element_by_id("MainContent_lnkOrderByAZ")#set order
wd.execute_script("arguments[0].click();", AtoZorder)
time.sleep(10)

adv_opt = wd.find_element_by_xpath('//*[@id="imgchkCType_0"]')#manufacturers
wd.execute_script("arguments[0].click();", adv_opt)
time.sleep(10)

##finding total number of pages
soup = BeautifulSoup(wd.page_source)
total_pages = soup.find_all(class_ = 'dvpagean')
total_pages = int(total_pages[-1].text)
print(total_pages)

#total_pages = 427
name_list = []
address_list  = []
dealer_of_list = []
for page_num in range(1,total_pages): 
  print("currently on: ", page_num) 
  num=0
  if (page_num<=5):
    num = 14
  else: 
    if (page_num<=total_pages-6):
      num=15
    else:
      if (page_num==total_pages-5): 
        num = 14 #####to be changed/html/body/form/div[4]/div/div[7]/div[3]/div[1]/div[4]/div[3]/div[4]/div[1]/div[12]/a
      else:
        if (page_num < total_pages-3):
          num=13
        else:
          num=12
  
  soup = BeautifulSoup(wd.page_source)

  list_1 = soup.find_all(class_ = "dv545mabgc")
  list_2 = soup.find_all(class_ = "dv535mawbgcnb")

  for l in range(len(list_1)):
    name = list_1[l].find(class_ ="gvHeadingtxt dvmtwlh").text
    address = list_2[l].find(class_ ="dv18lh").text
    address = address.replace('\n', "")
    address = address[0:-10]
    address = ' '.join(address.split())
    name = ' '.join(name.split())
    dealer_of = list_2[l].find_all('a')
    dealer_of = dealer_of[2:-12]
    for j in range(len(dealer_of)):
      dealer_of[j] = dealer_of[j].text
    name_list.append(name)
    address_list.append(address)
    dealer_of_list.append(dealer_of)

  if page_num != total_pages-1:  
    next= wd.find_element_by_xpath('/html/body/form/div[4]/div/div[7]/div[3]/div[1]/div[4]/div[3]/div[4]/div[1]/div[' + str(num) +']/a')   #next button
    scroll_element_into_view(wd, next)
    wd.execute_script("arguments[0].click();", next)
    time.sleep(20)
    wd.save_screenshot('SS'+str(l)+'.png')

print(len(name_list), len(address_list), len(dealer_of_list))



```

```
rows = zip(name_list, address_list, dealer_of_list)   ###writing to csv
import csv

with open('data.csv', "w") as f:
    writer = csv.writer(f)
    writer.writerow(('name', 'address', 'product'))
    for row in rows:
        writer.writerow(row)

print(len(address_list))
```
